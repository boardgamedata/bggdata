```python
import requests as rq
import xml.etree.ElementTree as ET
import time
import json
```

```python
uri = 'http://www.boardgamegeek.com/xmlapi2/collection?username=Aecios&stats=1&rated=1'
bgg = rq.get(uri)
while (bgg.status_code == 202):
    time.sleep(10)
    bgg = rq.get(uri)
```

```python
mycol = bgg.text

```

```python
collection = ET.fromstring(mycol)
```

```python
notas_dic = {}
for juego in collection.findall('item'):
    #rating = juego.find("rating")
    nombre = juego.find("name").text
    nota = juego.find("stats").find("rating").get("value")
    #print(str(nombre) + ": " + str(nota))
    notas_dic[nombre] = nota
    #print (rating.get("value"))

for n in notas_dic:
    print(n + str(notas_dic[n]))
    
```
