```python
import requests as rq
import xml.etree.ElementTree as ET
```

```python
bgg = rq.get('http://www.boardgamegeek.com/xmlapi2/collection?username=roblax&stats=1')
```

```python
print(bgg.status_code)
print(bgg.text)
```

```python
mycol = bgg.text

```

```python
collection = ET.fromstring(mycol)
print(collection.tag)
```

```python
for juego in collection:
    for caracteristica in juego:
        print(caracteristica.tag)
        print(caracteristica.attrib)
        print(caracteristica.text)
```
